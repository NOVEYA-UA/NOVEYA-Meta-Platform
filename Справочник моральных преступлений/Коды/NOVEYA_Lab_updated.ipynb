```python
# 📄 Загрузка и отображение файлов концепции и проектов
import docx

def read_docx(file_path):
    doc = docx.Document(file_path)
    return '\n'.join([p.text for p in doc.paragraphs if p.text.strip()])

concept_path = '/content/drive/MyDrive/NOVEYA/Аннотация Единой Методологии НОВЕЯ.docx'
projects_path = '/content/drive/MyDrive/NOVEYA/ПРОЕКТЫ НОВЕЯ.docx'

concept_text = read_docx(concept_path)
projects_text = read_docx(projects_path)

print('--- Концепция НОВЕЯ ---')
print(concept_text[:2000])
print('\n--- Проекты НОВЕЯ ---')
print(projects_text[:2000])
```
